.wav to Text Conversion

```python
from pydub import AudioSegment
from pydub.utils import make_chunks
import speech_recognition as sr

def process_audio(file_name):
    txtf = open("the_audio.txt", "w+")
    myaudio = AudioSegment.from_wav(file_name)
    chunks_length_ms = 8000
    chunks = make_chunks(myaudio, chunks_length_ms)
    for i, chunks in enumerate(chunks):
        chunkName = './chunked/' + file_name + "_{0}.wav".format(i)
        print('I am exporting' , chunkName)
        chunks.export(chunkName, format="wav")
        file = chunkName
        r = sr.Recognizer()
        with sr.AudioFile(file) as source:
            audio_listened = r.listen(source)
        try:
            rec = r.recognize_google(audio_listened)
            txtf.write(rec+".\n")
        except sr.UnknownValueError:
            print("I don't recognize your audio")
        except sr.RequestError as e:
            print("Could not get the result, Check your connection")


try:
    os.makedirs("chunked")
except:
    pass

process_audio("second.wav")
```

Text Summarizing using the extracted text

```python
import bs4 as bs
import urllib.request
import re
import nltk
```

```python
with open(f'the_audio.txt') as f:
    article = f.readlines()

article_text = " ".join(article)

print(article_text)
```

```python
# Removing Square Brackets and Extra Spaces
article_text = re.sub(r'\[[0-9]*\]', ' ', article_text)
article_text = re.sub(r'\s+', ' ', article_text)

print(article_text)
```

```python
# Removing special characters and digits
nltk.download('punkt')
nltk.download('stopwords')
formatted_article_text = re.sub('[^a-zA-Z]', ' ', article_text )
formatted_article_text = re.sub(r'\s+', ' ', formatted_article_text)
sentence_list = nltk.sent_tokenize(article_text)
stopwords = nltk.corpus.stopwords.words('english')

print(sentence_list)
```

```python
word_frequencies = {}
for word in nltk.word_tokenize(formatted_article_text):
    if word not in stopwords:
        if word not in word_frequencies.keys():
            word_frequencies[word] = 1
        else:
            word_frequencies[word] += 1
    maximum_frequncy = max(word_frequencies.values(), default=0)
for word in word_frequencies.keys():
    word_frequencies[word] = (word_frequencies[word]/maximum_frequncy)
    sentence_scores = {}
    
print(word_frequencies)
```

```python
#Sentence_Scoring
for sent in sentence_list:
    for word in nltk.word_tokenize(sent.lower()):
        if word in word_frequencies.keys():
            if len(sent.split(' ')) < 30:
                if sent not in sentence_scores.keys():
                    sentence_scores[sent] = word_frequencies[word]
                else:
                    sentence_scores[sent] += word_frequencies[word]
                    
print(sentence_scores)
```

Getting Final Summary in textual and audio format

```python
import heapq
summary_sentences = heapq.nlargest(7, sentence_scores, key=sentence_scores.get)

summary = ' '.join(summary_sentences)
print(summary)

from gtts import gTTS #Import Google Text to Speech
from IPython.display import Audio #Import Audio method from IPython's Display Class
tts = gTTS(summary) #Provide the string to convert to speech
tts.save('1.wav') #save the string converted to speech as a .wav file
sound_file = '1.wav'
Audio(sound_file, autoplay=True)
```
